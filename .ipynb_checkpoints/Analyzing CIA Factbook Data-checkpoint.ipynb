# Analyzing CIA Factbook Data Using SQLite and Python

In this project we are going to analyze demograpic data of all countries and we will answer following questions:
- Which countries have the highest population density?
- Which countries have the highest ratios of water to land? Which countries have more water than land?
- Which countries have the highest and which the lowest ratio of death rate to birth rate? Which countries have the highest migration rate?

We'll use data from the CIA World Factbook (https://www.cia.gov/library/publications/the-world-factbook/). The data contains demographic information about all of the countries:

- **id** - unique id of each row.
- **code** - country code.
- **name** - name of a country.
- **area** - whole area of a country in squared km.
- **area_land** - land area of a country in squared km.
- **area_water** - water area of a country in squared km.
- **population** - The population as of 2015.
- **population_growth** - The annual population growth rate, as a percentage.
- **birth_rate** - The number of births a year per 1000 people.
- **death_rate** - The number of death a year per 1000 people.
- **migration_rate** - The difference between the number of immigrants and the number of emigrants.
- **created_at** - date of creating the information.
- **updated_at** - date of last update of the information.

## Connecting to database

```python
#Let's import the libraries and connect to database
import sqlite3
import pandas as pd

conn = sqlite3.connect(r"Data\factbook.db")
q= "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(q,conn)
```

```python
q1 = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(q1,conn)
```

## Summary Statistics

```python
q2 = "SELECT MAX(population), MIN(population), MAX(population_growth), MIN(population_growth) FROM facts"
pd.read_sql_query(q2, conn)
```

From statistics above we can observe that there are countries, which population is equal to 0 and over 7.2 billion people. Let's check this outliers.

## Outliers

```python
q3 = "SELECT * FROM facts WHERE population = 0"
pd.read_sql_query(q3, conn)
```

```python
q4 = "SELECT * FROM facts WHERE population > 7000000000"
pd.read_sql_query(q4, conn)
```

It looks like 0 population is for Antarctics and over 7.2 billion for the world, which make sense. Let's create histograms for population, population_growth, birth_rate and death_rate (excluding this two outliers).

## Histograms

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
q5 = "SELECT population, population_growth, birth_rate, death_rate FROM facts WHERE population != (select min(population) from facts) AND population != (select max(population) from facts)"
pd.read_sql_query(q5, conn).hist(ax=ax)
plt.show()
```

## Which countries have the highest population density?

```python
q6 = "SELECT name, MAX(cast(population as float) / cast(area as float)) AS max_density, population, area FROM facts"
pd.read_sql_query(q6, conn)
```

```python
q8 = "select name, cast(population as float)/cast(area as float) AS density, population, area from facts order by density desc limit 10"
pd.read_sql_query(q8, conn)
```

The country with the lowest density is Macau.

## Which countries have the highest ratios of water to land? Which countries have more water than land?

```python
q9 = "SELECT name, cast(area_water as float)/cast(area_land as float) AS water_land_ratio, area_water, area_land FROM facts ORDER BY water_land_ratio DESC LIMIT 10"
pd.read_sql_query(q9, conn)
```

There are two countries with more water than land, which are: British Indian Ocean Territory and Virgin Islands.

## Which countries have the highest and which the lowest ratio of death rate to birth rate? Which countries have the highest migration rate?

```python
q10 = "SELECT name, cast(death_rate as float)/cast(birth_rate as float) AS ratio, birth_rate, death_rate FROM facts ORDER BY ratio DESC LIMIT 20"
pd.read_sql_query(q10,conn)
```

```python
q11 = "SELECT name, cast(death_rate as float)/cast(birth_rate as float) AS ratio, birth_rate, death_rate FROM facts WHERE ratio IS NOT NULL ORDER BY ratio ASC LIMIT 20"
pd.read_sql_query(q11,conn)
```

```python
q12 = "SELECT name, migration_rate FROM facts ORDER BY migration_rate DESC LIMIT 20"
pd.read_sql_query(q12,conn)
```
